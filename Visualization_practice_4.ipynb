# Matplotlib

* mtcars data comes from 1974 Motor trend Magazine.
* This data includes fuel consumption and other ten aspects of car design for then current car models.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
mtcars = pd.read_csv('mtcars.csv')
mtcars.head()
```

```python
mtcars.shape
```

```python
plt.figure(figsize=(15,8))
plt.hist(mtcars['mpg'])
plt.title('MPG Distribution',size=20)
plt.xlabel('mpg')
plt.ylabel('frequency')
```

```python
mtcars['gear'].value_counts().plot(kind='pie',explode=[0.01,0.01,0.01])
```

```python
mtcars['gear'].value_counts().plot(kind='hist')
```

```python
#bar plot
pd.crosstab(index=mtcars['gear'],columns=mtcars['cyl'])
```

```python
# bar plot between 2 different categories 
pd.crosstab(index=mtcars['gear'],columns=mtcars['cyl']).plot(kind='bar')
#They designed majority of the 3 gear vechicles with 8cyl.
```

```python
plt.scatter(mtcars['gear'],mtcars['cyl'],edgecolors='red')
```

```python
# Graphical Representation of data
# import matplotlib.pyplot as plt
# Histogram
plt.hist(mtcars['mpg'])
```

```python
#creates histogram with 6bins and colours filled in it.
plt.hist(mtcars['mpg'],color='violet',edgecolor='red',bins=6) 
```

```python
mtcars.describe()
```

```python
#Boxplot
plt.boxplot(mtcars['cyl'])
```

```python
plt.figure(figsize=(10,))
plt.boxplot(mtcars['mpg'],vert=False)
plt.xlabel('Box Plot',size=15)
plt.ylabel('MPG',size=15)
plt.title('MPG Boxplot',size=20)
```

```python
#Violin Plot
#help(plt.violinplot)
plt.violinplot(mtcars["cyl"],)
```

# Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=pd.read_csv('tips.csv')
```

```python
tips.head()
```

```python
tips['day'].unique()
```

# Univariate  one variable data visualization

```python
#strip plot
plt.figure(figsize=(10,6))
sns.stripplot(y= 'tip', data=tips,jitter=True) #spread , no x label
plt.ylabel('tip ($)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x=tips['day'],y=tips['tip']) # No.of dots gives no of persons who has given tips
plt.title('Day Vs Tip',size=20 )
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x=tips['day'],y=tips['tip'],size=5,jitter=False) # No.of dots gives no of persons who has given tips
plt.title('Day Vs Tip',size=20 )
```

```python
#swarm plot
sns.swarmplot(x=tips['day'],y=tips['tip']) # more no of people gives tips
plt.title('Day Vs Tip',size=20)
```

```python
#swarm plot
sns.swarmplot(x=tips['day'],y=tips['tip'],hue=tips['sex']) # Sat and sun -Female giving more tips
plt.title('Day Vs Tip',size=20)
```

```python
#Change orientation
#swarm plot
sns.swarmplot(x=tips['day'],y=tips['tip'],hue=tips['sex'],orient='h') # Sat and sun -Female giving more tips
plt.title('Day Vs Tip',size=20)
plt.show()
```

# Box and Violin Plot

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=tips['day'],y=tips['tip']) 
plt.title('Day Vs Tip',size=20)
```

```python
#violin plot
plt.figure(figsize=(10,6))
sns.violinplot(x=tips['day'],y=tips['tip']) 
plt.title('Day Vs Tip',size=20)
```

```python
plt.subplot(1,2,1)
sns.boxplot(x=tips['day'],y=tips['tip']) 
plt.title('Day Vs Tip',size=20)
plt.subplot(1,2,2)
sns.violinplot(x=tips['day'],y=tips['tip']) 
plt.title('Day Vs Tip',size=20)
```

```python
#Combining Plots

sns.stripplot(x=tips['day'],y=tips['tip'],size=5,jitter=True)
sns.violinplot(x=tips['day'],y=tips['tip'],inner=None,color='grey')
plt.show()
```

# Bivariate  two variables data visualization

```python
#Join Plot

sns.jointplot(x=tips['total_bill'],y=tips['tip']) # Most of the tips are between 2 to 4 if the tot bill is between 10 to 20.
plt.show()
```

```python
#Density Plot

sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='hex')
plt.show()
```

```python
#Pair Plot

sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips,hue='sex') #color coding
plt.show()
```
