```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
euro=pd.read_csv('Euro_2012_stats_TEAM.csv')
euro
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Goals',y=None,hue_order='Goals',data=euro)
plt.xlabel('Team')
plt.ylabel('Goals')
```

```python
plt.hist(x='Goals',data=euro)
```

```python
plt.pie(x=euro['Goals'],labels=euro['Team'],data=euro)
```
