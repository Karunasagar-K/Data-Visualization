```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
month = ['Jan','Feb','Mar','Apr','May','June','July']
sales = [18000,10000,35000,30000,70000,65000,40000]
print(month)
print(sales)
```

```python
plt.figure(figsize=(10,6))
plt.title('MONTH WISE SALES',fontsize=20,color='r')
plt.plot(month,sales,color='g')
plt.xlabel('Month',fontsize=15,color='b')
plt.ylabel('Sales',fontsize=15,color='b')
plt.show()
```

```python
#Q-1  #  hp wise average mpg

mpg = round(cars.groupby('hp')['mpg'].mean().sort_values(ascending=False).reset_index(name='mpg'))
mpg
```

```python
#Q-1  #  hp wise average mpg

wt = cars.groupby('wt')['mpg'].mean().sort_values(ascending=False).reset_index(name='mpg')
wt
```

```python
gear = cars.groupby(by='gear')['mpg'].mean().reset_index(name='mpg')
gear
```

```python
cyl  = cars.groupby(by='cyl')['mpg'].mean().reset_index(name='mpg')
cyl
```

```python
carb = cars.groupby(by='carb')['mpg'].mean().reset_index(name='mpg')
carb
```
