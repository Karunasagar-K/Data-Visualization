```python
import pandas as pd
mtcars=pd.read_csv('/content/drive/MyDrive/Data science/Data Science Materials 15 September 21/September_29 Data Vizualization/mtcars.csv')
mtcars
```

### give 2 insights
#### 1.what is the hp wrt gear and cyl 

```python
round(pd.pivot_table(data=mtcars,values='hp',index='gear',columns='cyl',aggfunc='mean'))
```

* Hp is increasing wrt gear and cyl
* there is no data available on 8 cyl and 4 gear

### 2.what is the disp wrt to gear and cyl?

```python
round(pd.pivot_table(data=mtcars,
    values='disp',
    index='gear',
    columns='cyl',
aggfunc='mean'))
```
