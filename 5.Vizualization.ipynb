```python
days=[1,2,3,4,5,6]
no_of_product_sold = [10,12,23,45,68,89]
plt.plot(days,no_of_product_sold)
```

```python
days=[1,2,3,4,5,6]
no_of_product_sold = [10,12,23,45,68,89]
plt.plot(days,no_of_product_sold)
plt.title('Sales per Day')
plt.xlabel('Days')
plt.ylabel('No.of.Products')
plt.grid()
plt.show()
```

```python
Month = ['Jan','Feb','March','Apr','May','June']
EB_Units_Consumed = [65,70,79,80,71,80]
plt.plot(Month,EB_Units_Consumed)
plt.title('EB units per month')
plt.xlabel('Month')
plt.ylabel('Units Consumed')
plt.grid()
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.relplot(df,x='total_bill',y='tip')
```

```python
sns.relplot(df,x='total_bill',y='tip',hue='sex')
```

```python
sns.relplot(df,x='total_bill',y='tip',hue='sex',size='size')
```

```python
sns.relplot(df,x='total_bill',y='tip',hue='sex',size='size',style='smoker')
```

```python
sns.relplot(df,x='total_bill',y='tip',hue='sex',size='size',style='smoker',row='day',col='time')
```

```python
sns.boxplot(df['total_bill'])
```

```python
sns.violinplot(df['total_bill'])
```

```python
sns.jointplot(df,x='total_bill',y='tip')
```

```python
sns.pairplot(df)
```
