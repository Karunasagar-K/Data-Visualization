```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
auto = pd.read_csv('Automobile.csv')
auto.head()
```

## Initial analysis

```python
auto.shape
```

```python
auto.isna().sum()
```

## Univariate analysis

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.distplot(auto['city_mpg'])
plt.show()
```

```python
sns.distplot(auto['normalized_losses'],kde=False,rug=True) #kde graph is removed but rug is true nothing but frwquency of distribution
plt.show()
```

## Bivariate analysis

```python
sns.jointplot(auto['engine_size'], auto['horsepower'])
plt.show()
```

```python
#kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
sns.jointplot(auto['engine_size'], auto['horsepower'],kind='reg')
plt.show()
```

```python
sns.pairplot(auto[['normalized_losses','engine_size','horsepower']])
```

```python
sns.stripplot(auto['fuel_type'],auto['horsepower'])
```

```python
sns.swarmplot(auto['fuel_type'],auto['horsepower'])
```

```python
sns.boxplot(auto['number_of_doors'],auto['horsepower'],hue=auto['fuel_type'])
```

```python
plt.figure(figsize=(10,8))
sns.barplot(auto['body_style'],auto['horsepower'],hue=auto['engine_location'])
```

```python
plt.figure(figsize=(10,6))
sns.countplot(auto['body_style'],hue=auto['engine_location'])
```

```python
plt.figure(figsize=(10,8))
sns.pointplot(auto['fuel_system'],auto['horsepower'],hue=auto['number_of_doors'])
```

```python
sns.factorplot(x='fuel_type',y='horsepower',hue='number_of_doors',col='engine_location',data=auto,kind='box')
#Various kind of inputs are :{point,bar,count,box,violin,strip}
```

```python
sns.lmplot(x='horsepower',y='peak_rpm',data=auto, hue='fuel_type')
```
